## 1\. Preços `outliers` de diamante

```python
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```python
# gráfico do exercício 1

with sns.axes_style ('whitegrid'):

    chart = sns.boxplot(x=data["price"])
    chart.set(title='Segmentação de preço de diamentes por tipo de corte', xlabel='Preço em USD', ylabel="Tipo de corte");
    chart.figure.set_size_inches(w=20, h=7)
```

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```python
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```python
# gráfico do exercício 2


with sns.axes_style ('whitegrid'):

    chart = sns.boxplot(x=data["price"], y=data["cut"])
    chart.set(title='Segmentação de preço de diamentes por tipo de corte', xlabel='Preço em USD', ylabel="Tipo de corte");
    chart.figure.set_size_inches(w=20, h=7)
```

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```python
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```python
data.describe()
```

```python
# gráfico do exercício 3

with sns.axes_style('whitegrid'):

  grafico = sns.histplot(data, x=data['weight'], bins=15, kde=True)
  grafico.set(title='Segmentação do peso dos carros', xlabel='Peso', ylabel='Contagem');
```
