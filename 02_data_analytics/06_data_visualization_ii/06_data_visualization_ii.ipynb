## 1\. Preços `outliers` de diamante

```python
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```python
# gráfico do exercício 1

with sns.axes_style ('whitegrid'):

    chart = sns.boxplot(x=data["price"])
    chart.set(title='Segmentação de preço de diamentes por tipo de corte', xlabel='Preço em USD', ylabel="Tipo de corte");
    chart.figure.set_size_inches(w=20, h=7)
```

**Insight do gráfico 1**: A partir do gráfico, podemos afirmar que diamantes vendidos a partir de $12000 podem ser considerados outliers, independente do corte. 

Isso nos mostra que, mesmo tendo uma vasta variedade de cortes, existe um teto nos valores de venda.

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```python
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```python
# gráfico do exercício 2


with sns.axes_style('whitegrid'):

  grafico = sns.scatterplot(data=data, x="carat", y="price", hue="clarity", palette="pastel")
  grafico.set(title='Correlação entre o  preço do diamante e seu peso', xlabel='Peso (quilates)', ylabel='Preço (USD)');
  grafico.get_legend().set_title("Transparência");
  grafico.figure.set_size_inches(w=14, h=7)
```

**Insight do gráfico 2**: Podemos ver claramente uma correlação positiva entre peso (quilates) e preço. Quanto maior o peso, maior o preço. Porém, não temos apenas a variável de peso x preço, a transparência – ou pureza – também conta. Então quanto mais puro o diamente, menor é a volatilidade no valor.

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```python
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```python
data.describe()
```

```python
# gráfico do exercício 3

with sns.axes_style('whitegrid'):

  grafico = sns.histplot(data, x=data['weight'], bins=15, kde=True)
  grafico.set(title='Segmentação do peso dos carros', xlabel='Peso', ylabel='Contagem');
```

**Insight do gráfico 3**: Neste histograma podemos afirmar que há grande concentração de veículos pesando entre 2000 e 3000 lbs. Ao revisitar a base de dados, podemos encontrar veículos como Chevrolet Chevelle Malibu 307 pesando aproximadamente 3500 lbs, logo podemos afirmar que a produção se concentram em carros leves ou de peso médio.
