## 1\. Preço do diamante por tipo de corte

```python
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```python
# gráfico do exercício 1

diamonds = sns.barplot(data=data, x="cut", y="price", palette="bright")
diamonds.set_title("Preço médio do diamante de acordo com o corte", fontsize=14, fontweight="bold")
diamonds.set(xlabel="Tipo de corte", ylabel="Média de preço (USD)")
diamonds.figure.set_size_inches(10,6)
```

**Insight do gráfico 1**: A constatação que nos chama mais a atenção é que, cortes do tipo Fair (menor qualidade) e Premium possuem os dois maiores volumes de precificação. Isso nos mostra que o tipo de corte pode influenciar na precificação, mas não é o fator determinante. Há fatores ainda desconhecidos quem o valor de diamantes com corte do tipo Fair.

Cortes intermediários, como Good e Very Good possuem uma média parecida e sofrem menos volatilidade na precificação.

## 2\. Número de passageiros em dezembro por ano

```python
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```python
# gráfico do exercício 2

flights = data.query("month == 'Dec'")
flights
```

```python
pass_dezembro = sns.lineplot(data=flights, x="year", y="passengers", palette="bright")
pass_dezembro.set_title("Número de passageiros por ano no mês de dezembro", fontsize=14, fontweight="bold");
pass_dezembro.set(xlabel="Ano", ylabel="Passageiros")
pass_dezembro.figure.set_size_inches(10,6)
```

**Insight do gráfico 2**: O ano de 1956 apresentou um crescimento de 100%, ou duas vezes em relação a 1950. Apesar de uma ligeira queda em meados de 1959, há uma tendência do alta para os próximos anos.

## 3\. Numero de passageiros por mês entre 1949 e 1959

```python
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```python
# gráfico do exercício 3

flights = data.query("1949 <= year < 1960")
flights
```

```python
  pass_meses = sns.lineplot(data=flights, x="month", y="passengers", hue='year', palette="bright")
  pass_meses.set_title("Número de passageiros por mês entre 1949 e 1959", fontsize=16, fontweight="bold");
  pass_meses.set(xlabel="Mês", ylabel="Passageiros")
  pass_meses.get_legend().set_title("Ano");
  pass_meses.figure.set_size_inches(10,6)
```

**Insight do gráfico 3**: Podemos observar que cada ano supera os indicadores do ano interior, além do gráfico nos apresentar boa consistência entre os meses de cada ano.

Os meses com o maior número de passageiros variam entre junho e agosto, iniciando sua queda no número de passageiros em setembro. Além disso, em dezembro observamos outro aumento repentino de passageiros. Esta sazonalidade se faz presente em todos os anos e pode ser um ponto importante para projetar um planejamento de viagens, já que é sabido que o meio e o fim do ano possuem mais movimento do que os outros meses.
