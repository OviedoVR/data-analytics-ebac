## 2\. Números

Preencha as células de código para preencher os valores de (A), (B) e (C) na tabela de ticket médio abaixo:

<br>

| Dia   | Valor Total Vendas | Qtd Total Vendas | Ticket Medio |
|-------|--------------------|------------------|--------------|
| 19/01 | (A)                | 3                | 320.52       |
| 20/01 | 834.47             | (B)              | 119.21       |
| 23/01 | 15378.12           | 5                | (C)          |

```python
# (A)

qt_total_vendas_A = 3
ticket_medio_A = 320.52
valor_total_A = ticket_medio_A * qt_total_vendas_A
print(f'O valor total é R${valor_total_A}')


# Revisão
ticket_medio_A = valor_total_A / qt_total_vendas_A
print(f'O ticket médio é de R${ticket_medio_A}')
```

```python
# (B)
valor_total_B = 834.47
ticket_medio_B = 119.21

qt_total_vendas_B = valor_total_B / ticket_medio_B
print(f'A quantidade de vendas e {qt_total_vendas_B}')

# Revisão
ticket_medio_B = valor_total_B / qt_total_vendas_B
print(f'O ticket medio e de R${ticket_medio_B}')
```

```python
# (C)
valor_total_C = 15378.12
qt_total_vendas_C = 5
ticket_medio_C = valor_total_C / qt_total_vendas_C
print(f'O ticket medio e de R${ticket_medio_C}')
```

## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```python
#maiusculo
cancao = 'Roda mundo, roda gigante, roda moinho, roda piao.'
print(cancao.upper())
```

```python
#minusculo
cancao = 'Roda mundo, roda gigante, roda moinho, roda piao.'
print(cancao.lower())
```

```python
#titulo
cancao = 'Roda mundo, roda gigante, roda moinho, roda piao.'
print(cancao.title())
```
